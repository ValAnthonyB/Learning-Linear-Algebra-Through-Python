```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
```

# Visualizing 2-D and 3-D Vectors

### 2-D vectors

```python
v = np.array([1,3])
w = np.array([2,3])
v_plus_w = v + w
origin = [0], [0] # origin point

plt.figure()
plt.quiver(*origin, v[0], v[1], scale=17, color='r', label='v')
plt.quiver(*origin, w[0], w[1], scale=17, color='b',label='w')
plt.quiver(*origin, v_plus_w[0], v_plus_w[1], scale=17, color='g',label='v + w')
plt.legend()
plt.show()
```

### 3-D vectors

```python
a = np.array([0, 2, 1])
b = np.array([2, 2, 0])
a_plus_b = a + b

origin = [0], [0], [0]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(*origin, a[0], a[1], a[2], color='red', label='a')
ax.quiver(*origin, b[0], b[1], b[2], color='blue',label='b')
ax.quiver(*origin, a_plus_b[0], a_plus_b[1], a_plus_b[2], color='green',label='a + b')
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])
plt.legend()
plt.show()
```

### Lenth of a vector

```python
v = np.array([1,3,2])
w = np.array([4,1,7])
```

```python
def vector_length(v):
    length = 0
    for i in v:
        length += i**2

    return length
```

```python
vector_length(v)
```

# Dot Product

### From scratch

```python
def dot_product(v,w):
    ans = 0
    for i in range(0, len(v)):
        ans += v[i] * w[i]
        
    return ans
```

```python
dot_product(v,w)
```

### Numpy Linalg

```python
np.dot(v,w)
```

## Dot Product of a Matrix and a vector

```python
A = np.array([[1,0,0],[-1,1,0],[0,-1,1]])
x = np.array([4,1,7])
```

```python
def matrix_vector_product(A, x):
    m, n = A.shape
    b = np.zeros(n)
    
    for i in range(0, m):
        sum_ = 0
        for j in range(0, n):
            sum_ += A[i][j] * x[j]        
        b[i] = sum_
    
    return b
```

```python
matrix_vector_product(A, x)
```
